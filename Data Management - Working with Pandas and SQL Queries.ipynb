## 2. Write a query to get current Product list (Product ID and name).



### Using SQL

```python
# Selecting current products, then accessing their ProductID, ProductName from Products table
current_products = pd.read_sql_query("select ProductID, ProductName from Products WHERE Discontinued == 0", conn)
display(current_products)
```

### Using PANDAS

```python
products_df[products_df['Discontinued'] == '0'][['ProductID', 'ProductName']]
```

##3. Write a query to get discontinued Product list (Product ID and name).


### Using SQL

```python
# Selecting discontinued products, then accessing their ProductID, ProductName from Products table
discontinued_products = pd.read_sql_query("select ProductID, ProductName from Products WHERE Discontinued == 1", conn)

display(discontinued_products)
```

### Using PANDAS

```python
products_df[products_df['Discontinued'] == '1'][['ProductID', 'ProductName']]
```

##4. Write a query to get most expensive and least expensive Product list (name and unit price).


### Using SQL

```python
least_expensive_prod = pd.read_sql_query("SELECT ProductName, min(UnitPrice) FROM Products", conn)
most_expensive_prod = pd.read_sql_query("SELECT ProductName, max(UnitPrice) FROM Products", conn)

display(least_expensive_prod)
display(most_expensive_prod)
```

### Using Pandas

```python
max = products_df['UnitPrice'].max()
min = products_df['UnitPrice'].min()

display(products_df[(products_df['UnitPrice'] == max) | (products_df['UnitPrice'] == min) ][['ProductName', 'UnitPrice']])
```

## 5. Write a query to get Product list (id, name, unit price) where current products cost less than 20.


### Using SQL

```python
products_less20 = pd.read_sql_query("SELECT ProductID, ProductName, UnitPrice FROM Products WHERE UnitPrice < 20 and Discontinued == 0" , conn)
display(products_less20)
```

### Using Pandas

```python
products_df[(products_df['UnitPrice'] < 20) & (products_df['Discontinued'] == '0')][['ProductID', 'ProductName', 'UnitPrice']]
```
